```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2
import os
import skimage

import warnings
warnings.filterwarnings('ignore')

from keras.datasets import mnist
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Dropout
import re
```

```python
# Load data

nums = re.compile(r'(\d+)')
def key(value):
    parts = nums.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts

data1 = pd.read_csv("/home/ojas/sop_qm9/potential_energy.csv")
# data2 = pd.read_csv("/home/nahush/Desktop/nnfl-assignment-i/test_data_file_id.csv")
# t = cv2.imread('./data/final_train/682.jpg')

trainingImages1 = []

# for i in trainingImages:
#     j = skimage.transform.rescale(i, 0.25, anti_aliasing = True)
#     trainingImages1.append(j.astype('float32') / 255.)

#test_path = './data/final_test'

counter = 0

for i in sorted(os.listdir('/home/ojas/sop_qm9/images/'),key=key):
    imgFile='/home/ojas/sop_qm9/images/'+i
    img = cv2.imread(imgFile)
    img = cv2.resize(img,(64,64))
    print(type(img))
    img = img.astype('float32')
    trainingImages1.append(img/255.)
    counter = counter + 1
    print(i + " " +str(counter))

y_train = data1['potential_energy'].to_numpy() 

```

```python
print(y_train)
```

```python
#Load data part2

x_train = np.asarray(trainingImages1)

x_train.shape, y_train.shape
```

```python
# Sample image



# x_samp, y_samp = x_train[0], y_train[0]

# print ("Shape of x_samp =", x_samp.shape)
# plt.imshow(x_samp)
# plt.show()
```

```python
EPOCHS = 10
# NUM_CLASSES = 2 # (0, 1)
```

```python

# One hot encoding
# y_train = np_utils.to_categorical(y_train)

x_train.shape, y_train.shape
```

```python
from sklearn.model_selection import train_test_split

x, x_test, y, y_test = train_test_split(x_train, y_train, test_size=0.05, random_state=42)
print(x.shape, y.shape)
print(x_test.shape, y_test.shape)
```

```python

model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=(64,64,3), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(0.2))
model.add(Conv2D(64, (3, 3),activation='relu'))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(0.2))
model.add(Conv2D(128, (5, 5), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='linear'))
None
```

```python
from keras import optimizers
opt = optimizers.Adam(lr = 1e-3, decay = 1e-3 / 200)
model.compile(loss="mean_absolute_percentage_error", optimizer=opt)
```

```python
model.summary()
```

```python
model.fit(x, y, validation_split = 0.1, validation_data = (x_test, y_test), epochs=EPOCHS, batch_size=40)
```

```python
model.fit(x_train, y_train, validation_split = 0.1, validation_data = None, epochs=5, batch_size=40)
```

```python
model.save('model.h5')
```
